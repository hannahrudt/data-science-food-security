### Further Understanding Food Insecurity Metrics
**Food Insecurity**

Food insecurity describes a consistent lack of resources for and access to nutritious, culturally relevant, sufficient food. Hunger is a physical feeling that is experienced by all humans and does not consider the implications of an individual's food environment.

**Supplemental Nutrition Assistance Program**

The Supplemental Nutrition Assistance Program (SNAP, formerly known as “Food Stamps”) is the federal food assistance program. The number of SNAP recipients in a county reflects the resource considerations of food insecurity--can people pay for sufficient food? While the federal government spends over $65 billion for SNAP benefits each year, the benefits are physically distributed by the state governments. SNAP benefits are allotted monthly based on income and household size, and allow recipients to buy almost all grocery items, including fresh produce, proteins, grains and cereals, processed snacks, and non-alcoholic beverages. Recipients are not allowed to purchase alcoholic beverages or hot prepared food with SNAP benefits.

**Grocery Store Ratio**

Comparing the number of grocery stores in a given area to the number of convenience stores (bodegas, delis, corner stores, etc.) is an effective way to understand food environments. Food environments reflect the accessibility considerations of food insecurity--can people find sufficient food? Convenience stores provide little to no fresh food, so individuals living in areas with a lower grocery store to convenience store ratio have less physical access to appropriate and nutritious food.

# Data description

### USDA Food Environment Atlas 
This dataset is a large dataset that includes various food environment factors such as store, restaurant proximity, food prices, food, and nutrition assistance programs, and community characteristics, which affect people's food choices and diet quality. We chose several variables as indicators for food insecurity status which include: grocery stores per 1000 people, convenience stores per 1000 people, specialized food stores per 1000 people, SNAP participants as a percentage of the population, and SNAP benefits per capita. This data was created to provide an overview of the community's accessibility to healthy food and stimulate the research on factors influencing food choices and diet quality. These parts of the dataset were provided by USDA's Food and Nutrition Service, U.S. Census Bureau, and USDA's Agricultural Marketing Service. This dataset is indirectly related to people. It does not contain any identifying information of an individual. However, it can identify the population in a specific county, which might be connected to determining socioeconomic status. For instance, the population of SNAP recipients can indicate the socioeconomic status of the county, which can potentially be sensitive to some populations.

The dataset is displayed as a map and available as an excel file. 

This data is related to people but does not disclose any specific information about an individual. It can be considered sensitive because it shows the percent of the population who achieved higher education, so some people who live in a county with lower education attainment might feel anxiety.

## ANOVA Multiple Linear Regression Models
ANOVA or Analysis of Variance is a Test that can determine if specific predictors are significant in the model. It helps determine if these predictors are needed or add information given that other predictors are in the model.

While the above heatmaps are useful to get an intuitive feel for which features are have the strongest relation to Covid-19 vaccination rates, ANOVA testing will allow us to ruggedly choose which predictors to use in multivariable linear regression models.

For the first ANOVA test, each hypothesis and alternative corresponding to each row in the ANOVA table is stated for your convenience. For the rest of the tests, the conclusions will be simply stated. 

---

```python
show_anova_table('pct_covid_vax')
```

#### Comprehensive List of all Tests from Above Table

---
##### 1st Row: gdp
$H_o$: all predictor slopes are equal to 0: $\beta_{gdp}$ = 0. These predictors are not useful or needed in the model to predict Covid-19 vaccination data. 

$H_A$:  $\beta_{gdp} \neq$ 0. `gdp` is needed in the model

**Conclusion:** Because the p-value for this test is larger than $\alpha$ = 0.01, 3.262323e-01, null hypothesis can not be rejected. Hence, `gdp` does not add to the model given the intercept is there. 

---
##### 2nd Row: pct_ge_bach
$H_o$: all predictor slopes are equal to 0:  $\beta_{gdp}$= $\beta_{pct ge bach}$ =  0. These predictors are not useful or needed in the model to predict Covid-19 vaccination data. 

$H_A$:  $\beta_{gdp}$ or $\beta_{pct ge bach} \neq$ 0. `pct_ge_bach` is needed in the model

**Conclusion:** Because the p-value for this test is so small, about 2e-12, null hypothesis is rejected. Hence, `pct_ge_bach` adds to the model given that `gdp` is there.

# Limitations

Non-citizens like undocumented immigrants or those on specific visas are not eligible for SNAP benefits so their food insecurity would not be evident in this dataset. Additionally, eligible immigrants are less likely to use food stamps than eligible citizens. These factors make SNAP data an underestimate for food insecurity.

Furthermore, although SNAP data is a good indicator for food insecurity status in different counties, it is not the only measure. We tried to account for this limited variable by including the grocery store ratio, but there are other indicators of food insecurity that are not utilized and may affect our representation of food insecurity.

Undocumented immigrants may fear answering the US Census out of fear and distrust. It can be more difficult to count the homeless population in the US Census. These factors can underestimate the population in some areas.

Due to rows with NaN or infinite values for certain columns, not every county in the United States is included. Our final dataset has rows for about 2,700 counties, out of the 3,006 existing counties. Additionally, there are a few boroughs and census areas that are left out. In particular, as of the creation of the dataset, all of Texas and New Mexico are not included, and parts of California. 

Because the datasets were compiled from different government organizations and each dataset is updated at a different frequency, the timeframe for all the datasets are not aligned. Especially any data related to COVID-19 pandemic, such as the COVID-19 vaccination rate, is updated regularly and most up to date, however, the most recent updates for other datasets such as socioeconomic factors are only up to 2019. As there could have been significant changes between before and after the COVID-19 pandemic, this discrepancy might have affected our analyses.
